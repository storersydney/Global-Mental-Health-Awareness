## Exploring Relationships

```python
# Create a correlation matrix using pandas

df_numeric = df.select_dtypes(include=['number'])
df_numeric.corr()
```

### Heatmap

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create a correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_numeric.corr(), annot = True, ax = ax)
plt.show(corr)
```

The mental disorders that have the highest correlation to DALYs (Disability-affected Life Years) are Schizophrenia, Bipolar disorder, Eating disorders, and Anxiety disorders. 

Mental disorders that have the strongest correlations to one another in terms of percent of the population are as follows: 
* Bipolar and Eating disorders 
* Drug use disorders and Schizophrenia
* Anxiety disorders and Bipolar disorders
* Anxiety disorders and Eating disorders
* Drug use disorders and Eating disorders

### Scatterplots

```python
# Create a scatterplot to explore the relationship between DALYs for all mental disorders and the prevalence of eating disorders

DALYs_Eating_disorders = sns.lmplot(
    x='DALYs - Mental disorders (Percent)', 
    y='Prevalence - Eating disorders (Percent)', 
    data=df
)
plt.show(DALYs_Eating_disorders)
```

This chart shows a strong positive correlation between the variables.

```python
# Create a scatterplot in seaborn showing the relationship between the prevalence of eating disorders and bipolar disorders

Bipolar_Eating_Disorders = sns.lmplot(x = 'Prevalence - Bipolar disorder (Percent)', y = 'Prevalence - Eating disorders (Percent)', data = df)
plt.show(Bipolar_Eating_Disorders)
```

This plot shows another strong positive correlation between the variables. It would be interesting to determine if there is a comorbidity between these two diagnoses.

### Pair Plot

```python
# Create a pair plot of the dataset
g = sns.pairplot(df_numeric)

# Rotate the axis labels
if g.axes is not None:
    for ax in g.axes.flatten():
        ax.set_xlabel(ax.get_xlabel(), rotation=45, fontsize=10)  # Rotate x-axis labels
        ax.set_ylabel(ax.get_ylabel(), rotation=60, fontsize=10, labelpad=40)  # Rotate y-axis labels

plt.show()
```

It looks like the correlations mentioned above are in fact the strongest and I would like to explore them further. There is a linear relationship between the variables with the strongest correlations.
No real correlation appears to exist with DALYs and drug use disorders, depressive disorders, and alcohol use disorders. This is surprising for depressive disorders because of the strong relationships with DALYs and other mental disorder categories. Drug use disorders and alcohol use disorders may not have a direct relationship because this measurement of DALYs only measures the Disability-Affected Life Years for Mental Disorders, rather than substance-use disorders.

### Categorical Plot

```python
# Create a histogram showing the frequency of the percent of DALYs (burden of disease)

hist = sns.histplot(df['DALYs - Mental disorders (Percent)'], bins = 20, kde = True)
plt.show()
```

```python
# Create a new variable showing ranges of the burden of disease

df.loc[df['DALYs - Mental disorders (Percent)'] < 3, 'Burden of Disease - Mental disorders'] = 'Low'
```

```python
df.loc[(df['DALYs - Mental disorders (Percent)'] >= 3) & (df['DALYs - Mental disorders (Percent)'] < 7), 'Burden of Disease - Mental disorders'] = 'Medium'
```

```python
df.loc[df['DALYs - Mental disorders (Percent)'] >= 7, 'Burden of Disease - Mental disorders'] = 'High'
```

```python
df['Burden of Disease - Mental disorders'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Create a categorical plot in seaborn using the burden of disease categories assigned above

sns.set(style="ticks")
catplot_bod_anxiety_disorders = sns.catplot(x='Prevalence - Anxiety disorders (Percent)', y='DALYs - Mental disorders (Percent)', hue='Burden of Disease - Mental disorders', data=df, palette='Reds')
plt.show()
```

As the prevalence of anxiety disorders increases in a population, the burden of disease also generally increases. There is a positive linear relationship between the variables. This plot makes it easier to see the different levels of the burden of disease in relation to anxiety disorders and DALYs.

## Research Questions and Hypotheses

* Although we cannot be sure how exactly the disorders affect DALYs, we have defined some strong correlations between variables. The only mental disorder that does not show a clear relationship with DALYs is Depressive disorders.

How do substance use diagnoses affect DALYs?
* There is no clear relationship shown between substance-use disorders and DALYs for Mental disorders. This is probably due to categorization of mental vs substance use disorders when collecting data only for DALYs - Mental disorders.

Which disorders affect DALYs the most?
* The mental disorders that have the strongest correlation to DALYs (Disability-affected Life Years) are Schizophrenia, Bipolar disorder, Eating disorders, and Anxiety disorders.


#### Hypotheses

If the prevalence of certain mental disorders increases, the DALYs will increase.

Prevalence of Schizophrenaia, Bipolar disorder, Eating disorders, and Anxiety disorders causes a higher burden of disease.
